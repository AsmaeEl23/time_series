```python
import pandas as pd
from math import sin
from math import radians
from matplotlib import pyplot
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
# create a differenced series
def difference(dataset, interval=1): # interval => l'ordre de differenciation
 diff = list()
 for i in range(interval, len(dataset)):
     value = dataset[i] - dataset[i - interval]
     diff.append(value)
 return pd.Series(diff)
```

```python
# invert differenced forecast
def inverse_difference(last_ob, value):
 return value + last_ob

#def invert_difference(orig_data, diff_data, interval):
# return [diff_data[i-interval] + orig_data[i-interval] for i in range(interval, len(orig_data))]
```

```python
##### Differencing to Remove Trends
 
# define a dataset with a linear trend
data = [i+1 for i in range(20)]
print(data)
# difference the dataset
diff = difference(data)
print(diff)
print(list(diff))
# invert the difference
inverted = [inverse_difference(data[i], diff[i]) for i in range(len(diff))]
print(inverted)
```

```python

##### Differencing to Remove Seasonality

 
# define a dataset with seasonality
data = [sin(radians(i)) for i in range(360)] + [sin(radians(i)) for i in range(360)]
pyplot.plot(data)
pyplot.show()
```

```python
# difference the dataset
diff = difference(data, 360)
pyplot.plot(diff)
pyplot.show()
```

```python

# invert the difference
inverted = [inverse_difference(data[i], diff[i]) for i in range(len(diff))]
pyplot.plot(inverted)
pyplot.show()
```
